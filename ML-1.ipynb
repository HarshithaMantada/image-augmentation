```python
white_img = np.full((400,400),255,dtype=np.uint8)
black_img = np.zeros((400,400),dtype=np.uint8)
shades_img = np.full((400,400),200,dtype=np.uint8)
```

```python
cv2.imshow("white",white_img)
cv2.imshow("black",black_img)
cv2.imshow("shades",shades_img)

cv2.waitKey(1) # 0 and no values means infinite dealy to close X button

cv2.destroyAllWindows()
```
